## <a href="https://towardsdatascience.com/all-you-need-to-know-about-pca-technique-in-machine-learning-443b0c2be9a1" target="_blank">PCA</a> & Feature transformation

```python
df_cov_matrix = np.cov(X_train.T)
eig_vals, eig_vecs = np.linalg.eig(df_cov_matrix)
# print('Eigenvectors \n%s' %eig_vecs)
# print('\nEigenvalues \n%s' %eig_vals)
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
total = sum(eig_vals)
var_exp = [(i / total)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
print('Variance captured by each component is \n',var_exp)
print('Cumulative variance captured as we travel with each component \n',cum_var_exp)
```

```python
pca = PCA(n_components=17).fit(X_train)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('No of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

Widzimy, że już przy 17 zmiennych otrzymujemy 'Cumulative explained variance' na poziomie 99%, w związku z tym moglibyśmy zrezygnować z ostatnich kilku ostatnich zmiennych.  

```python
X_train_pca  = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
X_test_pca.shape
```

```python
#listy do wykresów
list_of_methods=['accuracy', 'balanced_accuracy', 'recall', 'precision', 'f_measure']
list_of_name=[]
list_of_result_accuracy=[]
list_of_result_balanced_accuracy=[]
list_of_result_recall=[]
list_of_result_precision=[]
list_of_result_f_measure=[]
list_of_metric=[list_of_result_accuracy,list_of_result_balanced_accuracy,list_of_result_recall,list_of_result_precision,list_of_result_f_measure]

# X_train = X_test_sc
# X_test = X_test_sc

for index, (name, classifier) in enumerate(classifiers.items()):
    print(name)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    plot_confusion_matrix(classifier, X_test, y_test)
    accuracy = accuracy_score(y_test, y_pred)
    balanced_accuracy = balanced_accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred, average='weighted')
    precision = precision_score(y_test, y_pred, average='weighted')
    f_measure = f1_score(y_test, y_pred, average='macro')
   
    #WYKRESY I WSZYSTKO Z NIMI ZWIAZANE
    
    #listy do wykresów
    list_of_name.append(name)
    list_of_result_accuracy.append(accuracy)
    list_of_result_balanced_accuracy.append( balanced_accuracy)
    list_of_result_recall.append(recall)
    list_of_result_precision.append(precision)
    list_of_result_f_measure.append(f_measure)

    # Make a dataset:
    height = [accuracy, balanced_accuracy, recall, precision, f_measure]
    bars = ('accuracy', 'balanced_accuracy', 'recall', 'precision', 'f_measure')
    y_pos = np.arange(len(height))
 
    #Figsize
    plt.figure(figsize=(10,5))
 
    # Create bars
    plt.bar(y_pos, height, color = 'red')
    plt.ylim([0,1])
    # Create names on the x-axis
    plt.xticks(y_pos, bars)
    plt.xlabel('sposób oceny', fontsize=12, color='#323232')
    plt.ylabel('DOKŁADNOŚĆ', fontsize=12, color='#323232')
    plt.title('PORÓWNANIE DLA '+name, fontsize=16, color='#323232')
    # Show graphic
    plt.show();
     
    #KONIEC RZECZY Z WYKRESAMI 
    
    
    #print("Accuracy (train) for %s: %0.1f%% " % (name, accuracy * 100))
    #print("Ballanced Accuracy (train) for %s: %0.1f%% " % (name, balanced_accuracy * 100))
    #print("Recall (train) for %s: %0.1f%% " % (name, recall * 100))
    #print("Precision (train) for %s: %0.1f%% " % (name, precision * 100))
    #print("F-measure (train) for %s: %0.1f%% " % (name, f_measure * 100))

#Narysowanie wykresów które porównują tą samą metryke dla różnych metod klasyfikacji 
for x in range(len(list_of_metric)):
    height = [list_of_metric[x][y] for y in range(len(classifiers))]
    bars = (y for y in list_of_name )
    y_pos = np.arange(len(height))
 
    #Figsize
    plt.figure(figsize=(10,5))
 
    # Create bars
    plt.ylim([0.0,1.0])
    plt.bar(y_pos, height, color = 'red')
    # Create names on the x-axis
    plt.xticks(y_pos, bars)
    plt.xlabel('metoda', fontsize=12, color='#323232')
    plt.ylabel('DOKŁADNOŚĆ', fontsize=12, color='#323232')
    plt.title('Porównujemy '+list_of_methods[x], fontsize=16, color='#323232')
    # Show graphic
    plt.show();
```
